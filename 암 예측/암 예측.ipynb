```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#데이터 로드 
dataset = pd.read_csv('diabetes.csv', encoding='cp949')
dataset
```

```python
dataset.isnull().sum()
```

```python
# 데이터 분할
x_data = dataset.drop(['Outcome'], axis=1)
x_data
```

```python
y_data = dataset['Outcome']
y_data
```

```python
# 테스트셋 분할
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=1)
```

```python
# 모델 구축
model = keras.models.Sequential([keras.layers.Dense(20, input_dim = 8, activation='relu'),
                                keras.layers.Dense(8, activation='relu'),
                                keras.layers.Dense(1, activation='sigmoid'),
])
```

```python
# 모델 컴파일
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
# 모델 훈련
history = model.fit(x_train, y_train, epochs = 200, batch_size=50, validation_data=(x_test, y_test))
```

```python
# 학습 결과 그래프
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# 가상의 환자 데이터 입력
patient_1 = np.array([[3,173,78,39, 185,33.8,0.97,31]])

# 모델로 예측하기
prediction = model.predict(patient_1)

# 예측결과 출력하기
print(prediction)
```
